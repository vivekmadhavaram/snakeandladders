```python
from random import randint
```

```python
def snake(a):

        if(a==24):

             return 5

        if(a==43):

             return 38

        if(a==56):

             return 25

        if(a==61):

            return 59

        if(a==69):

            return 12

        if(a==86):

            return 54

        if(a==90):

            return 72

        if(a==96):

            return 84

        if(a==98):

            return 58

        else:

            return a
```

```python
def ladder(a):

        if(a==3):

            return 24

        if(a==8):

            return 55

        if(a==16):

            return 26
        if(a==32):

            return 48

        if(a==37):

            return 76

        if(a==50):

            return 70

        if(a==64):

            return 83

        if(a==77):

            return 95

        if(a==80):

            return 99

        if(a==89):

            return 91

        else:

            return a
```
